## Prepare data for visualizations

```python
SECRETS_PATH = os.path.join(os.getcwd(), "scripts", "configs", "secrets.env")
GROUP = "SEX&RAC1P"
```

```python
db_client = CoreDBClient(SECRETS_PATH)
db_client.connect()
```

```python
def get_best_pps_per_exp_config(exp_config_name):
    # Find uuids of the best pp per exp_config, lp, and run_num
    pipeline_query = get_best_pps_per_lp_and_run_num_query(exp_config_name)
    subquery = pipeline_query[:3] # We need just three first steps to get best pp uuids
    results = list(db_client.client[db_client.db_name][PHYSICAL_PIPELINE_OBSERVATIONS_TABLE].aggregate(subquery))
    
    return pd.json_normalize(results)
```
